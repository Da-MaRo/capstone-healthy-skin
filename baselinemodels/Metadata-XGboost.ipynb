### Hypothesis: Can we infer the malignancy of the lesion by patient information age,sex  and location of lesion.

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import xgboost as xgb
```

```python
# load csv
data = pd.read_csv('../data/processed/Metadata_processed.csv')
```

```python
data.head()
```

```python
# define X and y
y = data['dx_binary']
X = data.drop(['dx_binary'], axis=1)
X = X.drop(["lesion_id"], axis=1)
X = X.drop(["image_id"], axis=1)
X = X.drop(["dataset"], axis=1)
X = X.drop(['dx'], axis=1)
X = X.drop(['dx_type'], axis=1)
```

```python
X.head()
```

```python
y.head()
```

```python
# split data into train and test sets
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y)
```

```python
# one hot encode categorical variables
from sklearn.preprocessing import OneHotEncoder

encoder = OneHotEncoder(handle_unknown='ignore', drop='if_binary', sparse_output=False, categories='auto')

X_train = pd.DataFrame(encoder.fit_transform(X_train))
X_test = pd.DataFrame(encoder.transform(X_test))
```

```python
#get feature names

feature_names = list(encoder.get_feature_names_out())
```

```python
# encode labels
from sklearn.preprocessing import LabelEncoder

encoder = LabelEncoder()

y_train = encoder.fit_transform(y_train)
y_test = encoder.transform(y_test)

# view labels
y_train[:10]
```

```python
# train xgboost model
from xgboost import XGBClassifier

model = XGBClassifier(n_estimators=1000, learning_rate=0.05, n_jobs=4)

# fit model on training data
model.fit(X_train, y_train, eval_set=[(X_test, y_test)], verbose=False)
```

```python
# get the best model parameters from cross validation
from sklearn.model_selection import GridSearchCV

param_grid = {'n_estimators': [100, 200, 500, 1000, 2000, 5000, 10000],
                'max_depth': [2, 3, 4, 5, 6, 7, 8, 9],
                'learning_rate': [0.001,0.01, 0.05, 0.1]}
model = XGBClassifier(n_jobs=4)

# define grid search
grid_search = GridSearchCV(model, param_grid, scoring='accuracy', n_jobs=-1, verbose=1)
# execute grid search
grid_result = grid_search.fit(X_train, y_train)
# summarize results
print("Best: %f using %s" % (grid_result.best_score_, grid_result.best_params_))
```

```python
# cross validation with XGBoost

#from sklearn.model_selection import cross_validate

#cv_results = cross_validate(model, X_train, y_train, cv=5, scoring='accuracy', return_train_score=True, n_jobs=-1, verbose=1)
#cv_results['test_score']
```

```python
# use the best parameters to train the model
best_model = XGBClassifier(**grid_result.best_params_, n_jobs=4)
best_model.fit(X_train, y_train, eval_set=[(X_test, y_test)], verbose=False)
```

```python
# make predictions for test data
y_pred = best_model.predict(X_test)
```

```python
best_model.get_booster().feature_names = feature_names
```

```python
# evaluate predictions
from sklearn.metrics import accuracy_score

accuracy = accuracy_score(y_test, y_pred)
print("Accuracy: %.2f%%" % (accuracy * 100.0))

# plot feature importance
from xgboost import plot_importance

plot_importance(best_model)

# increase figure size
plt.rcParams['figure.figsize'] = [20, 10]
plt.show()
```

```python
# confusion matrix
from sklearn.metrics import confusion_matrix

cm = confusion_matrix(y_test, y_pred)
print(cm)
```

```python
# roc auc score
from sklearn.metrics import roc_auc_score

print(roc_auc_score(y_test, y_pred))

# roc curve
from sklearn.metrics import roc_curve

fpr, tpr, thresholds = roc_curve(y_test, y_pred)

plt.plot(fpr, tpr, color='orange', label='ROC')
plt.plot([0, 1], [0, 1], color='darkblue', linestyle='--', label='Random')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve Metadata-XGBoost')
```
